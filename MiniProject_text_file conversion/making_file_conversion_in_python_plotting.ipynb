```python
import random as r
import string as s

def make_file(name_, no_of_kb):
    fp=open("_"+name_+'.txt','w')	# Open the file in writing mode
    for i in range(0, no_of_kb):
        for  j in range(0,32):##1KB
            fp.write("".join(r.sample(s.ascii_letters, 32)))
    fp.close()

size = list(range(200,1200,200))
for i in size:
    make_file(str(i),i*1024)
```

```python
import time
```

```python
def convert_to_uppercase(file_path):
    with open(file_path, 'r') as file:
        content = file.read()
    upper_content = content.upper()
    with open(file_path, 'w') as file:
        file.write(upper_content)

time_taken_python = []
for i in size:
    start = time.time()
    convert_to_uppercase('_'+str(i)+'.txt')
    time_taken_python.append(time.time() - start)
time_taken_python
```

```python
import pandas as pd
```

```python
df = pd.read_csv('outputc.csv')
time_taken_c = []
time_taken_c.append(float(df.columns[0]))
for i in df.iloc[:,0]:
    time_taken_c.append(i)

time_taken_c
```

```python
df = pd.read_csv('output.csv')
time_taken_cpp = []
time_taken_cpp.append(float(df.columns[0])/1000000)
for i in df.iloc[:,0]:
    time_taken_cpp.append(i/1000000)

time_taken_cpp
```

```python
import matplotlib.pyplot as plt
import numpy as np
  
plt.plot(size, time_taken_python, label = "Python")
plt.plot(size, time_taken_cpp, label = "C++")
plt.plot(size, time_taken_c, label = "C")
plt.xlabel("File size in MB")
plt.ylabel("Time taken in seconds")
plt.legend()
plt.show()
```
